```python
import datetime
# Cadastra equipamentos novos
# Equipamentos contém status de locado ou livres
# Só possivel alugar se estiver disponível
# Remove equipamentos
# Mostra todos os disponíveis
# Bonus: Ao encerrar a execução, cria um txt com as locações e devoluções do dia.

#Abre a lista de equipamentos locados e não locados.
list = open('equipment.txt', 'w')
list.write("Prancha de Surf,R$500,LOCADO")
list.write("\nBola de futebol,R$55,DISPONIVEL")
list.close()
#Abre o Log de todas as informações de entrada e saída.
records = open('records.txt', 'w')
records.close()

#Função principal que chama o menu cliente ou funcionário.
def equipmentRental():
    print('''Olá seja bem-vindo à FasT, a melhor locadora de equipamentos esportivos do norte-nordeste.
    Primeiramente nos diga, você é:
    [1] Funcionário
    [2] Cliente''') 
    loginOptions(str(input('Digite o número ou a palavra correspondente:')))
    
#Função que faz a triagem da resposta do primeiro menu, checa se é funcionário ou cliente, e redireciona para o menu correto.
def loginOptions(userType):
    if 'FUNCIONÁRIO' in userType.upper() or 'FUNCIONARIO' in userType.upper() or '1' == userType:
        employeeMenu('Employee')
    elif 'CLIENTE' in userType.upper() or '2' == userType:
        userMenu('User')
    else:
        print('Opção inválida, por favor digite corretamente.')
        return equipmentRental()
            
#Menu com as opções de usuário comum.
def userMenu(userType):
    print('''Menu do usuário:
    [1] Mostrar Equipamentos
    [2] Alugar Equipamento
    [3] Devolver Equipamento
    [4] Voltar ao login
        ''')
    userOption = str(input('Digite o número da opção:'))
    if userOption == '1':
        equipmentList(userType)
    elif userOption == '2':
        turnRented()
    elif userOption == '3':
        returnEquipment()
    elif userOption == '4':
        return equipmentRental()
    else:
        print('Opção inválida.')
        return userMenu(userType)

#Menu com as opções de funcionário.
def employeeMenu(userType, authenticated = False):
    senha = '444555'
    if authenticated:
        print('''Menu do funcionário:
        [1] Mostrar Equipamentos
        [2] Remover Equipamentos
        [3] Cadastrar Equipamento
        [4] Mostrar Registros
        [5] Voltar ao login
        ''')
        employeeOption = str(input('Digite o número da opção:'))
        if employeeOption == '1':
            equipmentList(userType, authenticated)
        elif employeeOption == '2':
            removeEquipment(authenticated)
        elif employeeOption == '3':
            addEquipment(authenticated)
        elif employeeOption == '4':
            showRecords(authenticated)
        elif employeeOption == '5':
            return equipmentRental()
        else:
            print('Opção inválida.')
    else:
      employeePass = str(input('Digite a senha de funcionário, apenas números:'))
      if employeePass == senha:
        authenticated = True
      else:
        authenticated = False
        print('Senha inválida!')
    return employeeMenu(userType, authenticated)

# Função que retorna a lista de equipamentos organizada.
def equipmentList(userType, authenticated = False):
    # Mostra os produtos
    list = open('equipment.txt', 'r')
    index = 0;
    print('EQUIPAMENTOS'.center(50,'='))
    for item in list:
      itemsplit = item.replace("\n", "").split(',')
      print(f"[{index}] {itemsplit[0]} ({itemsplit[2]})")
      index = index + 1;
    print(''.center(50,'='))
    list.close()
    #Checa o UserType e volta ao menu correto.
    if userType == 'User':
      return userMenu(userType)
    elif userType == 'Employee':
      return employeeMenu(userType, authenticated)
    
# Função que adiciona equipamentos à lista de equipamentos.
def addEquipment(authenticated = False):
    loop = 'Y'
    while loop == 'Y':
        productsList = open('equipment.txt', 'a')
        itemName = str(input('Digite o nome do produto que deseja adicionar.'))
        itemPrice = str(input('Digite o preço do produto que deseja adicionar.'))
        productsList.write(f'\n{itemName},R${itemPrice},DISPONIVEL')
        print('Produto adicionado com sucesso.')
        takeRecord(itemName, 'Adicionado')
        loopAnswer = str(input('Quer adicionar mais produtos? [Y/N]'))
        loop = loopAnswer.upper()
        productsList.close()
    equipmentList('Employee', authenticated)

# Função que remove equipamentos da lista de equipamentos.
def removeEquipment(authenticated = False):
    loop = 'Y'
    while loop =='Y':
        productsList = open('equipment.txt', 'r')
        fullList = productsList.readlines()
        productsList.close()
        equipmentList('')
        indexRemoveItem = int(input('Digite o ID do produto à ser removido:'))
        RemoveItem = fullList.pop(indexRemoveItem).split(',')
        if RemoveItem[2] == 'DISPONIVEL':
          takeRecord(RemoveItem[0], 'Removido')
          productsList = open('equipment.txt', 'w')
          for item in fullList:
              productsList.write(item)
          print('Produto removido com sucesso, nova lista:')
          productsList.close()
        else:
          print('Produto não pode ser removido pois esta locado.')
        equipmentList('')
        loopAnswer = str(input('Quer remover mais produtos? [Y/N]'))
        loop = loopAnswer.upper()
    equipmentList('Employee', authenticated)

#Função que aluga os produtos para um usuário comum.
def turnRented():
    equipmentList('')
    rentID = int(input('Digite o ID do produto que quer alugar:'))
    openList = open('equipment.txt', 'r')
    productList = openList.readlines()
    openList.close()
    item = productList[rentID].replace("\n", "").split(',')
    if item[2] == 'DISPONIVEL': 
        takeRecord(item[0], 'Locado')
        productList[rentID] = productList[rentID].replace('DISPONIVEL', 'LOCADO')
        changeList = open('equipment.txt', 'w')
        for item in productList:
            changeList.write(item)
        changeList.close()
        print('Produto Alugado com sucesso.')
        equipmentList('User')
    else:
        print('''
        ======== Produto já alugado ==========
        ======== Tente outro produto =========
        ''')
        equipmentList('User')
        
# Função para devolução dos produtos do usuário comum.
def returnEquipment():
    equipmentList('')
    rentID = int(input('Digite o ID do produto que quer devolver:'))
    openList = open('equipment.txt', 'r')
    productList = openList.readlines()
    openList.close()
    item = productList[rentID].replace("\n", "").split(',')
    if item[2] == 'LOCADO':  
        takeRecord(item[0], 'Devolvido')
        productList[rentID] = productList[rentID].replace('LOCADO', 'DISPONIVEL')
        changeList = open('equipment.txt', 'w')
        for item in productList:
            changeList.write(item)
        changeList.close()
        print('Produto devolvido com sucesso!')
        equipmentList('User')
    else:
        print('''
        ======== Você não pode devolver ==========
        ========   Algo que não alugou! ==========
        ''')
        equipmentList('User')
    
# Função que salva os logs de registro/adição/devolução/locação.
def takeRecord(equipment, action):
    datenow = datetime.datetime.now()
    records = open('records.txt', 'a')
    records.write(f'{datenow:%d/%m/%Y %H:%M},{equipment},{action}\n')
    records.close()

# Função que puxa os registros e mostra na tela.
def showRecords(authenticated = False):
    records = open('records.txt', 'r')
    print('REGISTROS'.center(50,'='))
    for item in records:
      itemsplit = item.replace("\n", "").split(',')
      print(f"O {itemsplit[1]} foi {itemsplit[2]} em {itemsplit[0]}")
    print(''.center(50,'='))
    records.close()
    showRecordsRB = str(input('Quer voltar ao menu de funcionário? [Y]/[N]'))
    if showRecordsRB.upper() == 'Y':
        return employeeMenu('Employee', authenticated)
    elif showRecordsRB.upper() == 'N':
        print('Fazendo Logout...')
    else:
        print('Opção Inválida. Mostrando registros novamente.')
        return showRecords()
    
    
equipmentRental()
```
